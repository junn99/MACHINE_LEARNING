### 조건 조회
- 조건문의 결과는 T&F이므로, T나 F만 추출하는 형태로 주로 사용

```python
# 조건절(조건문)의 결과는 True, False
df['DistanceFromHome'] > 10
```

```python
df.loc[df['DistanceFromHome'] > 10]
```

```python
df.loc[(df['PercentSalaryHike'] < 20) & (df['TotalWorkingYears'] < 10)]
```

```python
df.loc[df['MonthlyIncome'] > 7000, ['Age']]
    # 조건에 맞는 특정열만 조회
```

### 기타 등등

```python
df.loc[df['JobSatisfaction'].isin([1, 4])]
    # 아래와 같음
```

```python
df.loc[(df['JobSatisfaction'] == 1) | (df['JobSatisfaction'] == 4)]
    # 위와 같음
```

```python
df.loc[df['Age'].between(18, 25)]
```

# 집계 메서드

```python
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Attrition_simple2.CSV'
df = pd.read_csv(path)
df.head()
```

```python
df["Age"].sum()
```

```python
df.groupby("Age",as_index=False)["Gender"].count().head()
    # 나이별 성별 수(= 사람 수)
    # as_index = True : 행 번호를 기반으로 한 정수값이 인덱스로 설정
    # as_index = Flase : 집계 기준이 되는 열이 인덱스 열됨
df.groupby("Gender",as_index=False)["Age"].count()
    # 성별별 나이의 개수(= 사람 수)
```
