#### native-country

```python
df['native-country'].value_counts()
```

- 수많은 종류가 있고, 특히 미국이 압도적 비중
- 어떻게 처리?
    1. 미국이 대다수이므로 나머지 국가를 묶기 : 그만큼 정보는 줄어듬
    2. 지역별로 묶기
    3. class 평균값에 대해 비슷한 국가끼리 묶기
- 제일 중요한 것은 종속변수에 대한 유사성 있어야 묶을 수 있음.

```python
# 국가별 평균 확인
df.groupby('native-country')[['age','education-num','capital-gain','capital-loss','hours-per-week','class']].mean().sort_values('class')
```

```python
for i in df.columns:
    if df[i].dtype == int:
        print(i)
```

- class의 대부분을 차지하는 미국의 값을 보았을 때 값이 평균값으로 0.24이다
- 다른 값들을 보면 묶을 수 있을 것 같진 않다.
- 지역별로 묶는 것도 아무튼 좋지 않다.
- 범주형 데이터를 무작정 숫자로 치환하여 모델링은? = 라벨링
    - 대부분 상황에선 좋지않지만, 트리 기반 모델에선 허용

```python
country_group = df.groupby('native-country')['class'].mean()
    # 그룹별 class의 평균값 저장
```

```python
country_group
```

```python
country_group = country_group.reset_index()
    # 행 인덱스를 변수(열)로 빼줌
```

```python
country_group
```

```python
df = df.merge(country_group, on='native-country',how='left')
```

```python
df
```

```python
df.drop('native-country', axis=1, inplace=True)
df=df.rename(columns={'class_x':'class','class_y':'native-country'})
```

```python
df
```

## 2.2 전처리 - 결측치 처리 & 더미 변수 변환

```python
df.isna().mean()
    # 결측치 비율 확인
    # 3개의 변수에서 확인됨
```

- native-country : 원랜 범주형 변수이므로 평군치로 채우는 게 불가능!
    - 지금은 숫자(class의 평균값)이므로 결측치 채울 수 있음 -> mean or 아무숫자(-9,-99)
    - 하지만 선형모델에선 주의

```python
df['native-country'] = df['native-country'].fillna(-99)
```

- 나머지 2개 변수는 모두 범주형 변수이므로 평균치로 해결 불가
    - 특정 텍스트를 채우거나 해당 행을 지움

```python
df['workclass'].value_counts()
```

workclass 변수를 보면 특정 값이 대부분을 차지함
- 이런 경우엔 해당 값으로 결측치 채우면 가능

```python
df['workclass'] = df['workclass'].fillna("Private")
```

```python
df['occupation'].value_counts()
```

occupation의 경우엔 애매하다
- 별도 텍스트 값으로 채움

```python
df["occupation"] = df["occupation"].fillna("Unknown")
```

```python
# 결측치 해결 다 했으면, 범주형 데이터를 더미 변수로 변환

df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

## 3. 모델링 & 평가

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('class', axis=1),
                                                    df['class'],test_size=0.4,random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
```
